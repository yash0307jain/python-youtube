#### 3. Conditional statements

1. if
2. elif
3. else


```python
age = 20

if age >= 18:
    print("You are an adult.")
elif age >= 13:
    print("You are a teenager.")
else:
    print("You are a child.")
```

---

#### 4. Loops

1. for
2. while


```python
# For loop
for i in range(5):  # [0, 5)
    print("Count:", i)
```

```python
# While loop
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```

#### 7. Data structure in Python

1. Lists
2. Tuples
3. Sets
4. Dictionaries


```python
# List (mutable)
fruits = ["apple", "banana", "mango"]
print(fruits)
```

```python
print(type(fruits))
```

```python
# How to append anything to the list
fruits.append("orange")
print(fruits)
```

```python
# Tuple (immutable)
numbers = (1, 2, 3, 4)
print(numbers)
```

```python
print(type(numbers))
```

```python
# Set
unique_nums = {1, 2, 2, 3, 4}
print(unique_nums)
```

```python
print(type(unique_nums))
```

```python
# How to append anything to the set
unique_nums.add(5)
print(unique_nums)
```

```python
# Dictionary (its like a json object)
person = {
    "name": "Yash",
    "experience_years": 6,
    "is_coder": True,
}
print(person)
```

```python
print(type(person))
```

```python
# How to call a specific key in dictionary
person_name = person["name"]
print(person_name)
```

```python
# How to add a new key in dictionary
person["hobby"] = "Table Tennis"
print(person)
```
