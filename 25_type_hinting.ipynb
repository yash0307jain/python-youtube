## Type Hinting and Annotations


#### 1. Basic function without type hints


```python
def add(a, b):
    return a + b


print(add(10, 5))
print(add("Hello", "World"))  # Works, but types are unclear
```

```python
print(add(10, "Yash"))
```

---

#### 2. Type hints added


```python
def add(a: int, b: int) -> int:
    return a + b


print(add(10, 5))
print(add("Hello", "World"))  # âŒ Type checker will show warning
```

#### 4. Function with Multiple Types


```python
from typing import Union


def process_data(data: Union[int, str, float]) -> str:
    return f"Processed {data}"


print(process_data(123))
print(process_data(1.23))
print(process_data("text"))
```

```python
def process_data(data: int | str | float) -> str:
    return f"Processed {data}"


print(process_data(123))
print(process_data(1.23))
print(process_data("text"))
```

#### 5. Lists, Tuples, and Dictionaries


```python
from typing import List, Tuple, Dict

print(list == List)
print(tuple == Tuple)
print(dict == Dict)
```

```python
from typing import List, Tuple, Dict


def display_items(items: List[str]) -> None:
    for item in items:
        print(item)


def get_user() -> Tuple[int, str]:
    return 101, "Yash"


def get_scores() -> Dict[str, int]:
    return {"Yash": 90, "Bob": 85}


# Example Usage
display_items(["Python", "Type", "Hints"])
print(get_user())
print(get_scores())
```

```python
int_value, str_value = get_user()
print(int_value, str_value)
```

#### 6. Optional Parameters


```python
from typing import Optional


def greet(name: Optional[str]) -> None:
    if name:
        print(f"Hello, {name}!")
    else:
        print("Hello, Guest!")


greet("Yash")
greet(None)
```

```python
from typing import Union


def greet(name: Union[str, None]) -> None:
    if name:
        print(f"Hello, {name}!")
    else:
        print("Hello, Guest!")


greet("Yash")
greet(None)
```

```python
def greet(name: str | None) -> None:
    if name:
        print(f"Hello, {name}!")
    else:
        print("Hello, Guest!")


greet("Yash")
greet(None)
```

#### 7. Type Aliases


```python
from typing import Dict, List

# Create alias for cleaner code
UserData = Dict[str, List[int]]


def show_user_stats(data: UserData) -> None:
    for user, scores in data.items():
        print(f"{user}: {scores}")


users = {"Yash": [10, 20, 30], "Bob": [15, 25, 35]}

show_user_stats(users)
```

#### 8. Generic Type Hints


```python
from typing import TypeVar, List

T = TypeVar("T")  # Declare a type variable


def get_first_item(items: List[T]) -> T:
    return items[0]


print(get_first_item([10, 20, 30]))
print(get_first_item(["a", "b", "c"]))
```

```python
var = get_first_item([10, 20, 30])
var2 = get_first_item(["a", "b", "c"])
var3 = get_first_item([1.0, 2.0, 3.0])
```

#### 9. Callable (Functions as Arguments)


```python
from typing import Callable


def apply_operation(a: int, b: int, operation: Callable[[int, int], int]) -> int:
    return operation(a, b)


def add(x: int, y: int) -> int:
    return x + y


def multiply(x: int, y: int) -> int:
    return x * y


print(apply_operation(5, 3, add))
print(apply_operation(5, 3, multiply))
```

#### 10. Any and NoReturn


```python
from typing import Any, NoReturn


def print_value(value: Any) -> None:
    print("Value:", value)


def fatal_error(message: str) -> NoReturn:
    raise SystemExit(message)


print_value(123)
print_value("Hello")
```

```python
fatal_error("Something went wrong!")  # exits program
```

#### 11. Type Hinting for Classes and Methods


```python
class Rectangle:
    def __init__(self, width: float, height: float) -> None:
        self.width = width
        self.height = height

    def area(self) -> float:
        return self.width * self.height


rect = Rectangle(10.0, 5.0)
print("Area:", rect.area())
```
