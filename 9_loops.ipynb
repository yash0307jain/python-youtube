## Loops

-   Loops are for repeating actions.


```python
# Without a loop:
print("Hello")
print("Hello")
print("Hello")
```

```python
# With a 'for' loop:
print("\nNow with a loop:")
for _ in range(5):
    print("Hello")
```

---

#### `for` loop

-   A `for` loop goes through each item in a sequence.


Looping through a `list`


```python
fruits = ["apple", "banana", "cherry", "orange"]

print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[3])
```

```python
fruits = ["apple", "banana", "cherry", "orange"]
print("--- My Fruit Basket ---")

for fruit in fruits:
    print(f"I have a {fruit}")

print("All fruits listed!")
```

Looping through `string`


```python
word = "PYTHON"
print("\n--- Spelling out PYTHON ---")

for letter in word:
    print(letter)
```

`range()` function

-   range() generates a sequence of numbers.


```python
# 1. range(stop) - Loops from 0 up to (but not including) 5
print("\n--- range(5) ---")
for num in range(5):
    print(f"Number: {num}")
```

```python
# 2. range(start, stop) - Loops from 2 up to (but not including) 6
print("\n--- range(2, 6) ---")
for num in range(2, 6):
    print(f"Number: {num}")
```

```python
# 3. range(start, stop, step) - Loops from 10 to 100, counting by 20
print("\n--- range(10, 101, 20) ---")
for num in range(10, 101, 20):
    print(f"Number: {num}")
```

#### `while` loop

-   A `while` loop repeats as long as its condition is true.


```python
# A simple countdown
print("\n--- Countdown ---")
count = 5

while count > 0:
    print(count)
    count = count - 1  # Or: count -= 1
    # This line is crucial! It prevents an infinite loop.

print("Liftoff! ðŸš€")
```

```python
# INFINITE LOOP

n = 1
while n > 0:
    print(f"Looping forever... n is {n}")
    n = n + 1  # n is always greater than 0, so this never stops
```

#### `break` Statement

-   To exit early


```python
numbers = [1, 5, 12, 19, 21, 30]
print(f"\nSearching for a number > 20 in {numbers}")

for num in numbers:
    print(f"Checking {num}...")
    if num > 20:
        print(f"Found it! {num} is greater than 20.")
        break  # Exit the loop now

print("Search over.")
```

#### `continue` Statement

-   To skip an iteration


```python
# Print only the ODD numbers
print("\n--- Printing only ODD numbers from 1 to 10 ---")
for num in range(1, 11):
    if num % 2 == 0:  # Check if the number is even
        continue  # If it's even, skip the print and go to the next number

    print(num)
```

#### `else` Clause in Loops

-   The `else` block runs when the loop finishes without a `break`.


```python
# Case 1: Loop completes (no break)
print("\n--- Checking list 1 for a 'banana' ---")
fruits = ["apple", "cherry", "orange"]
for fruit in fruits:
    print(f"Checking {fruit}...")
    if fruit == "banana":
        print("Found a banana!")
        break
else:
    # This will run because "banana" was not found and 'break' was not called
    print("No banana found in the list.")
```

```python
# Case 2: Loop is broken
print("\n--- Checking list 2 for a 'banana' ---")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"Checking {fruit}...")
    if fruit == "banana":
        print("Found a banana!")
        break
else:
    # This will NOT run because the loop was stopped by 'break'
    print("No banana found in the list.")
```
