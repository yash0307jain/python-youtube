#### Mutable and Immutable

Mutable

-   List
-   Set
-   Dictionary

Immutable

-   Integers
-   Float
-   String
-   Boolean
-   Tuple


```python
# mutable
my_list = [1, 2, 3]

print(id(my_list))
```

```python
my_list
```

```python
my_list.append(4)

print(id(my_list))
```
