#### Common string methods

-   upper()


```python
name = "Yash Jain"
print(name.upper())
```

-   lower()


```python
name = "Yash Jain"
print(name.lower())
```

-   title()


```python
name = "yash jain"
print(name.title())
```

-   find()


```python
name = "Yash Jain"
print(name.find("a"))
```

-   replace()


```python
name = "Yash Jai"
print(name.replace("Jai", "Jain"))
```

-   strip()


```python
name = "   Yash    Jain   "
print(name.strip())
```

-   split()


```python
name = "YashhelloJainhellocoder"
print(name.split("hello"))
```

-   startswith()


```python
name = "Yash Jain"
print(name.startswith("yash"))
```

-   endswith()


```python
name = "Yash Jain"
print(name.endswith("Jain"))
```

#### String concatenation and repetition


```python
first_name = "Yash"
last_name = "Jain"

full_name = first_name + " " + last_name
print(full_name)
```

```python
first_name = "Yash "
print(first_name * 5)
```

#### String formatting (f string and format)


```python
# format
template = "My name is {}"
print(template.format("Yash Jain"))
```

```python
template = "My name is {name} {hobby}"
print(template.format(name="Yash Jain", hobby="coder"))
```

```python
# f - string
name = "Yash Jain"
hobby = "coding"
template = f"My name is {name} and my hobby is {hobby}"

print(template)
```
