#### 1. Check if a Number is Prime


```python
num = 8
is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

Let say we have

n / d = k

n = d \* k

-   n is number
-   d is the divisible number
-   k is the quatient

For Example

-   if n = 36
-   square root of n = 6
-   there d and k cannot be greater than 6
-   for example if d = 7 and k = 8, then there product will be 56, which is greater
-   for example if d = 18 and k = 2, then there product will be 36, but in this case 18 is already a combination of 2 _ 3 _ 3
-   but if we take d = 6 and k = 6, then there product will be 36, which is equal

Therefore whenever we need to check the prime we check till `square root` of the given number


```python
# Optimize
num = 28
is_prime = True

for i in range(2, int(num**0.5) + 1):  # run till square root (floor) of that number
    if num % i == 0:
        is_prime = False
        break

if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

#### 2. Reverse a Number


```python
num = 12345
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed number:", rev)
```

#### 3. Sum of Digits for a number


```python
num = 12345
sum_digits = 0

while num != 0:
    digit = num % 10
    sum_digits += digit
    num //= 10

print("Sum of digits:", sum_digits)
```

#### 4. Find the Largest Element in a List


```python
numbers = [12, 45, 9, 38, 67, 23]
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest number:", largest)
```

```python
print("Largest number:", max(numbers))
```

#### 5. Fibonacci Series (N terms)

-   0, 1, 1, 2, 3, 5, 8, 13, 21...


```python
0 + 1 = 1
1 + 1 = 2
1 + 2 = 3
2 + 3 = 5
3 + 5 = 8
5 + 8 = 13
```

```python
num = 10
num_0, num_1 = 0, 1

print(num_0, end=" ")
for _ in range(num):
    temp = num_0 + num_1
    num_0 = num_1
    num_1 = temp
    print(num_0, end=" ")
```

#### 6. Count Vowels in a String


```python
text = "Loops make coding fun"
vowels = "aeiou"
count = 0

for ch in text:
    if ch.lower() in vowels:
        count += 1

print("Vowel count:", count)
```

#### 7. Display All Prime Numbers in a Range


```python
start, end = 10, 50

for num in range(start, end + 1):
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                break
        else:
            print(num, end=" ")
```

#### 9. Check if a String is a Palindrome


madam
madam


```python
text = "madamm"
rev = ""

for ch in text:
    rev = ch + rev

if text == rev:
    print(f"'{text}', '{rev}' is a palindrome")
else:
    print(f"'{text}', '{rev}' is not a palindrome")
```

```python
text = "madamm"
rev = text[::-1]

if text == rev:
    print(f"'{text}', '{rev}' is a palindrome")
else:
    print(f"'{text}', '{rev}' is not a palindrome")
```

#### 10. Check if a Number is an Armstrong Number


153 = 1^3 + 5^3 + 3^3
153 = 1 + 125 + 27
153 = 153


```python
num = 154
original_num = num
sum_power = 0

while num != 0:
    digit = num % 10
    sum_power += digit**3
    num //= 10

if sum_power == original_num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```
