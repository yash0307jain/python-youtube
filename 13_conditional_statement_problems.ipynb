## Conditional Statement Problems


#### 1. Positive, Negative, or Zero


```python
# Problem 1: Check if number is positive, negative or zero
num = int(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

#### 2. Even or Odd


```python
# Problem 2: Check if a number is even or odd
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number.")
else:
    print("Odd number.")
```

#### 3. Largest of Three Numbers


```python
# Problem 3: Find the largest among three numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    print(f"{a} is the largest.")
elif b >= a and b >= c:
    print(f"{b} is the largest.")
else:
    print(f"{c} is the largest.")
```

#### 4. Check Leap Year


```python
# Problem 4: Check if a year is a leap year
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

#### 5. Grade Calculator


```python
# Problem 5: Calculate grade based on marks
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
elif marks >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

#### 6. Check Vowel or Consonant


```python
# Problem 6: Check whether a character is vowel or consonant
ch = input("Enter a letter: ").lower()

# Check if input is a single alphabetic character
if len(ch) == 1 and ch.isalpha():
    if ch in ["a", "e", "i", "o", "u"]:
        print("Vowel")
    else:
        print("Consonant")
else:
    print(f"Invalid input! '{ch}', Please enter a single alphabetic character.")
```

#### 7. Age Category


```python
# Problem 7: Classify age group
age = int(input("Enter your age: "))

if age < 13:
    print("Child")
elif age < 20:
    print("Teenager")
elif age < 60:
    print("Adult")
else:
    print("Senior Citizen")
```

#### 8. Check Divisibility by 3 and 5


```python
# Problem 8: Divisible by 3, 5, both or neither (order of if elif is important)
num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print("Divisible by both 3 and 5")
elif num % 3 == 0:
    print("Divisible by 3 only")
elif num % 5 == 0:
    print("Divisible by 5 only")
else:
    print("Not divisible by 3 or 5")
```

#### 9. Simple Calculator


```python
# Problem 9: Simple calculator using if-elif-else
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operation (+, -, *, /): ")

if op == "+":
    print("Result:", num1 + num2)
elif op == "-":
    print("Result:", num1 - num2)
elif op == "*":
    print("Result:", num1 * num2)
elif op == "/":
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Cannot divide by zero!")
else:
    print("Invalid operator!")
```
