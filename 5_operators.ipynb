## Operators


#### Arithmetic Operators

-   Addtion
-   Subtraction
-   Multiplication
-   Division (float)
-   Floor Division
-   Module (remainder)
-   Exponentiation


```python
## Addtion
num1 = 10
num2 = 20

num3 = num1 + num2
print(num3)
```

```python
## Subtraction
num1 = 10
num2 = 20

num3 = num1 - num2
print(num3)
```

```python
## Multiplication
num1 = 10
num2 = 20

num3 = num1 * num2
print(num3)
```

```python
## Division float
num1 = 25
num2 = 20

num3 = num1 / num2
print(num3)
```

```python
## Division Floor
num1 = 45
num2 = 20

num3 = num1 // num2
print(num3)
```

```python
## Module
num1 = 41
num2 = 20

num3 = num1 % num2
print(num3)
```

```python
## Exponential
num1 = 5

num2 = num1**5
print(num2)
```

#### Comparison Operators

-   Equal to
-   Not equal
-   Greater than
-   Less than
-   Greater or equal
-   Less or equal


```python
## Equal to
num1 = 10
num2 = 20

print(num1 == num2)
```

```python
## Not Equal to
num1 = 10
num2 = 20

print(num1 != num2)
```

```python
## Greater than
num1 = 10
num2 = 20

print(num1 > num2)
```

```python
## Less than
num1 = 10
num2 = 20

print(num1 < num2)
```

```python
## Greater or equal to
num1 = 19
num2 = 20

print(num1 >= num2)
```

```python
## Less or equal to
num1 = 21
num2 = 20

print(num1 <= num2)
```

#### Logical Operators

-   and
-   or
-   not


```python
## and
num1 = 10
num2 = 20

result = num1 == 10 and num2 == 10
print(result)
```

```python
## or
num1 = 10
num2 = 20

result = num1 == 11 or num2 == 21 or 10 == 10
print(result)
```

```python
## not
num1 = 10
num2 = 20

result = not num1 == 11
print(result)
```

-   =
-   +=
-   -=
-   \*=
-   /=

-   %=

```python
# equal to
num1 = 10
```

```python
# +=
num1 += 20  # num1 = num1 + 20
print(num1)
```
