## String formatting


#### 1. String concatenation


```python
name = "Yash"
age = 25
height = 5.5
language = "Python"

result = (
    "My name is "
    + name
    + ". I am "
    + str(age)
    + " years old, my height is "
    + str(height)
    + " and I love "
    + language
    + "."
)
print(result)
```

---

#### 2. % Operator (Old Style Formatting)

-   `%s` → string placeholder
-   `%d` → integer placeholder
-   `%f` → float placeholder


```python
name = "Yash"
age = 25
height = 5.5
language = "Python"

result = "My name is %s. I am %d years old, my height is %f and I love %s." % (name, age, height, language)
print(result)
```

```python
# floating value - only 2 pointer value after decimal
result = "My name is %s. I am %d years old, my height is %.2f and I love %s." % (name, age, height, language)
print(result)
```

---

#### 3. str.format() Method


```python
name = "Yash"
age = 25
height = 5.5
language = "Python"

result = "My name is {}. I am {} years old, my height is {} and I love {}.".format(name, age, height, language)
print(result)
```

```python
# With indexed placeholders
print("My name is {0}. I am {1} years old, my height is {2} and I love {3}.".format(name, age, height, language))
```

```python
# With named placeholders
print(
    "My name is {n}. I am {a} years old, my height is {h} and I love {l}.".format(n=name, a=age, h=height, l=language)
)
```

---

#### 4. f-Strings (Literal String Interpolation)


```python
name = "Yash"
age = 25
height = 5.5
language = "Python"

result = f"My name is {name}. I am {age} years old, my height is {height} and I love {language}."
print(result)
```

```python
# You can also add expressions inside
print(f"In 5 years, I’ll be {age + 5} years old!")
```

---

#### 5. string.Template


```python
from string import Template

name = "Yash"
age = 25
height = 5.5
language = "Python"

template = Template("My name is $name. I am $age years old, my height is $height and I love $language.")
result = template.substitute(name=name, age=age, height=height, language=language)
print(result)
```
